```python
import pandas as pd
df = pd.read_csv('home_pricesMWR.csv')
df
```

```python
dummies = pd.get_dummies(df[['town']])
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['town', 'town_west_windsor'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop(['price'], axis='columns')
x
```

```python
y = final.price
y
```

```python
model.fit(x,y)
```

```python
import joblib
joblib.dump(model, 'modelMWR')
```

```python
MR = joblib.load('modelMWR')
```

```python
MR.predict([[3400,0,0]])
```

```python
MR.score(x,y)
```
